## Identificação do setor aeronáutico

### Subtask:
Carregar o arquivo `NCM_SH.csv` e identificar os códigos NCM (Nomenclatura Comum do Mercosul) que correspondem à cadeia produtiva do setor aeronáutico.


**Reasoning**:
Load the `NCM_SH.csv` file into a pandas DataFrame using the semicolon delimiter, then display the head and info to understand its structure and columns.

**Reasoning**:
The previous attempt to load the CSV failed due to a UnicodeDecodeError. This indicates that the file is not encoded in UTF-8. I will try loading the file again, specifying a different encoding, such as 'latin1', which is common for files with special characters.



```python
ncm_df = pd.read_csv('/content/NCM_SH.csv', delimiter=';', encoding='latin1')
display(ncm_df.head())
ncm_df.info()
```

**Reasoning**:
The `ncm_df` is loaded correctly. Now I need to identify the columns containing product descriptions and search for relevant keywords related to the aeronautical sector to extract the corresponding NCM codes. The columns `NO_SH6_POR`, `NO_SH4_POR`, `NO_SH2_POR` seem to contain product descriptions in Portuguese. I will search these columns for keywords like "aeronave", "aviação", "partes" (parts) and extract the `CO_SH6` codes.



```python
keywords = ['aeronave', 'aviação', 'partes']
aeronautical_ncm_codes = set()

for keyword in keywords:
    # Search in SH6 descriptions
    sh6_matches = ncm_df[ncm_df['NO_SH6_POR'].str.contains(keyword, case=False, na=False)]
    aeronautical_ncm_codes.update(sh6_matches['CO_SH6'].tolist())

    # Search in SH4 descriptions
    sh4_matches = ncm_df[ncm_df['NO_SH4_POR'].str.contains(keyword, case=False, na=False)]
    aeronautical_ncm_codes.update(sh4_matches['CO_SH4'].tolist())

    # Search in SH2 descriptions
    sh2_matches = ncm_df[ncm_df['NO_SH2_POR'].str.contains(keyword, case=False, na=False)]
    aeronautical_ncm_codes.update(sh2_matches['CO_SH2'].tolist())

print("Identified Aeronautical NCM Codes (SH6, SH4, SH2):")
print(aeronautical_ncm_codes)
```

## Visualização dos dados

### Subtask:
Criar visualizações para apresentar os resultados da análise de forma clara e compreensível.


**Reasoning**:
Check if the aeronautical_export_df DataFrame is empty and either create visualizations or print a message accordingly.



```python
import matplotlib.pyplot as plt
import seaborn as sns

if aeronautical_export_df.empty:
    print("No aeronautical export data available for visualization.")
else:
    print("Aeronautical export data is available. Generating visualizations.")

    # Convert VL_FOB to numeric
    aeronautical_export_df['VL_FOB'] = pd.to_numeric(aeronautical_export_df['VL_FOB'], errors='coerce')

    # Bar plot of total export value per year
    annual_export_value = aeronautical_export_df.groupby('CO_ANO')['VL_FOB'].sum().reset_index()
    plt.figure(figsize=(10, 6))
    sns.barplot(x='CO_ANO', y='VL_FOB', data=annual_export_value)
    plt.title('Total Aeronautical Export Value per Year')
    plt.xlabel('Year')
    plt.ylabel('Total Export Value (VL_FOB)')
    plt.show()

    # Bar plot of total export value per state (SG_UF_MUN) - if not too granular
    if aeronautical_export_df['SG_UF_MUN'].nunique() < 50: # Arbitrary threshold for granularity
        state_export_value = aeronautical_export_df.groupby('SG_UF_MUN')['VL_FOB'].sum().reset_index()
        plt.figure(figsize=(12, 6))
        sns.barplot(x='SG_UF_MUN', y='VL_FOB', data=state_export_value)
        plt.title('Total Aeronautical Export Value per State')
        plt.xlabel('State')
        plt.ylabel('Total Export Value (VL_FOB)')
        plt.show()
    else:
        print("Too many unique states to plot effectively.")

    # Table summarizing key statistics per year
    annual_summary = aeronautical_export_df.groupby('CO_ANO').agg(
        total_export_value=('VL_FOB', 'sum'),
        number_of_records=('CO_ANO', 'count')
    ).reset_index()
    print("\nAnnual Aeronautical Export Summary:")
    display(annual_summary)
```

## Unir dados de referência ao dataframe de exportação

### Subtask:
Unir os DataFrames de referência ao `aeronautical_export_df` usando as colunas de código apropriadas (SH4, CO_MUN, CO_PAIS).


**Reasoning**:
Perform left merges to add the names of product, municipality, and country to the aeronautical export data, then display the head and info of the merged dataframe.



```python
merged_df = pd.merge(aeronautical_export_df, ncm_mapping, how='left', left_on='SH4', right_on='CO_SH4')
merged_df = pd.merge(merged_df, mun_mapping, how='left', left_on='CO_MUN', right_on='CO_MUN_GEO')
merged_df = pd.merge(merged_df, pais_mapping, how='left', left_on='CO_PAIS', right_on='CO_PAIS')

display(merged_df.head())
merged_df.info()
```
