**Reasoning**:
Inside the loop, read the current CSV file into a pandas DataFrame, handling potential encoding and delimiter issues. Then display the head and columns of the loaded DataFrame to verify.



```python
export_files = ["/content/EXP_2021_MUN.csv", "/content/EXP_2022_MUN.csv", "/content/EXP_2023_MUN.csv", "/content/EXP_2024_MUN.csv", "/content/EXP_2025_MUN.csv"]

for file_path in export_files:
    print(f"Processing file: {file_path}")
    try:
        # Attempt to read with default parameters first
        df_export = pd.read_csv(file_path)
    except UnicodeDecodeError:
        # If UnicodeDecodeError occurs, try with 'latin1' encoding
        try:
            df_export = pd.read_csv(file_path, encoding='latin1')
        except:
            # If 'latin1' fails, try with 'ISO-8859-1'
            df_export = pd.read_csv(file_path, encoding='ISO-8859-1')
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")
        continue

    # Check if the data is in a single column and if it's delimited by semicolon
    if df_export.shape[1] == 1 and df_export.columns[0].count(';') > 0:
        try:
            df_export = pd.read_csv(file_path, encoding='latin1', sep=';')
        except:
             df_export = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';')

    print(f"Head of {file_path}:")
    display(df_export.head())
    print(f"Columns of {file_path}:")
    print(df_export.columns)
```

All the export files have been successfully loaded into DataFrames within the loop. The next step is to filter each of these DataFrames using the `sh4_codes` extracted earlier and then concatenate the filtered DataFrames into a single DataFrame.



```python
filtered_dfs = []

for file_path in export_files:
    print(f"Processing file: {file_path}")
    try:
        df_export = pd.read_csv(file_path, encoding='latin1', sep=';')
    except:
         df_export = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';')

    df_exp_aeronautico_single_file = df_export[df_export['SH4'].isin(sh4_codes)]
    filtered_dfs.append(df_exp_aeronautico_single_file)

df_exp_aeronautico = pd.concat(filtered_dfs, ignore_index=True)
display(df_exp_aeronautico.head())
print(df_exp_aeronautico.columns)
```

## Realizar cruzamento de dados (região)

### Subtask:
Cruzar o DataFrame de exportação com nomes com o DataFrame de municípios com região para adicionar a informação da região.

Merge the export data DataFrame with the municipalities DataFrame to add the region information based on the municipality name. Display the head of the resulting DataFrame.

I need to inspect the unique municipality names in both DataFrames to understand the discrepancies and potentially clean or standardize them before merging.



```python
print("Unique municipalities in df_exp_aeronautico_sp_nomes:")
print(df_exp_aeronautico_sp_nomes['NO_MUN'].unique())

print("\nUnique municipalities in df_municipios:")
print(df_municipios['Município'].unique())
```

**Reasoning**:
The municipality names have different casing and some have accented characters in `df_municipios` which might be causing the merge to fail. I will standardize the municipality names in both dataframes by converting them to uppercase and removing accented characters before attempting the merge again.



```python
import unicodedata

def remove_accents(input_str):
    if isinstance(input_str, str):
        nfkd_form = unicodedata.normalize('NFKD', input_str)
        return "".join([c for c in nfkd_form if not unicodedata.combining(c)])
    return input_str

df_exp_aeronautico_sp_nomes['NO_MUN_cleaned'] = df_exp_aeronautico_sp_nomes['NO_MUN'].apply(lambda x: remove_accents(x).upper())
df_municipios['Município_cleaned'] = df_municipios['Município'].apply(lambda x: remove_accents(x).upper())

df_exp_aeronautico_sp_com_regiao = pd.merge(
    df_exp_aeronautico_sp_nomes,
    df_municipios[['Região', 'Município_cleaned']],
    left_on='NO_MUN_cleaned',
    right_on='Município_cleaned',
    how='left'
)

# Drop the redundant cleaned columns and the original 'Município' column from df_municipios
df_exp_aeronautico_sp_com_regiao = df_exp_aeronautico_sp_com_regiao.drop(['NO_MUN_cleaned', 'Município_cleaned'], axis=1)

display(df_exp_aeronautico_sp_com_regiao.head())
```
